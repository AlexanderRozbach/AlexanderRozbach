```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/econometrica/main/econometrica-1/datasets/diamonds.csv')
```

```python
df
```

```python
df["price"]
```

```python
import matplotlib.pyplot as plt
```

```python
df.loc[: , ['table','price']]
```

```python
X = df.loc[:,['table']]
X
```

```python
Y = df.loc[:,['price']]
X = df.loc[:,['table']]
```

```python
plt.figure(figsize=(10,10))
plt.plot(X,Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_predict = model.predict(X)
y_predict
```

```python
plt.figure(figsize=(10,5))
plt.plot(X,Y,'o')
plt.plot(X,y_predict,'k-')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(Y,y_predict)
```

```python
r2_score(Y,y_predict)
```
